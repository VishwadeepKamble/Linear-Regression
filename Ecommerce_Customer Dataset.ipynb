Q11) Splitting data into train and test

```python
X = customers[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
y = customers['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

Q12) Import Linear Regression from sklearn.linear_model

```python
from sklearn.linear_model import LinearRegression
```

Q13) Create an instance of a LinearRegression() model and store it in variable lm.

```python
lm = LinearRegression()
```

Q14) Train/fit lm on the training data

```python
lm.fit(X_train,y_train)
```

Q15) Print out the coefficients of the mode

```python
print(pd.DataFrame(lm.coef_, X.columns, columns = ["Coefficients"])) 
```

Q16) evaluate its performance by predicting off the test values!

```python
Y_predicted =lm.predict(X_test)
Y_predicted
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
R2 = r2_score(y_test, Y_predicted)
Rmse = np.sqrt(mean_squared_error(y_test, Y_predicted))

print(f"R2: {R2}, Rmse: {Rmse}")
```

#### As we can see that accuracy is quite good of model

Q17) Create a scatterplot of the real test values versus the predicted values

```python
scatter = sns.scatterplot(y_test, Y_predicted)
plt.xlabel("Test values")
plt.ylabel("Predicted values")
plt.show(scatter)
```

Q18) Plot histogram of residuals and make sure it looks normally distributed.

```python
plt.hist((y_test - Y_predicted))
plt.show()
```
