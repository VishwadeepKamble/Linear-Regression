#### Imputing function to convert "Value" & "Wage" into "float"

```python
def str2float(euros):
    
    # convert M in value columns to Millions
    if euros [-1] == "M":                          # last index of (variable column)
        return float(euros[1:-1]) * 1000000        # converting to Millions
    elif euros [-1] == "K":
        return float(euros[1:-1]) * 1000           # converting to Thousands
    else:
        return float(euros[1:])
```

```python
# applying function on Value & Wage

df["Value"] = df["Value"].apply(lambda x: str2float(x))

df["Wage"] = df["Wage"].apply(lambda x: str2float(x))
```

#### Imputing function to convert "weight" into "float" datatype

```python
# 1 Pound (lbs) =  0.45359237 kilogram (kg)

def str_2_float(weight):
    # convert lbs in value columns to pounds
    if weight [-1] == "s":
        return round(float(weight[0:3])*0.45359237, 2)
    else:
        return float(weight[0:])
```

```python
# applying function on "Weight"

df["Weight"] = df["Weight"].apply(lambda x: str_2_float(x))
```

#### Replacing the "Nan" values of "Release Clause" with most frequent category in respective column with help of mode

```python
def impute_nan(df,variable):
    most_frequent_category = df[variable].mode()[0]
    df[variable].fillna(most_frequent_category, inplace = True)
```

```python
# replacing the "null" values of "Release Clause" with most frequent value

impute_nan(df, "Release Clause")
df.head()
```

#### Converting Height into "float" with help of imputing function

```python
def s2f(height):
    
    # replace ("'") with (".")
    
    if height [1] == "'":
        new_height = height.replace("'", ".")
        new_height_1 = round(pd.to_numeric(new_height),1)
        return new_height_1
    else:
        return float(height[1:])
```

```python
# applying s2f function on Height

df["Height"] = df["Height"].apply(lambda x: s2f(x))
```

### Perfroming Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 111)
```

```python
# Building function

def metrics(y_actual, y_pred):
    r2 = r2_score(y_actual, y_pred)
    rmse = np.sqrt(mean_squared_error(y_actual, y_pred))
    print("The R2 value is {} and RMSE value is {}".format(round(r2,2), round(rmse,2)))
```

### Vanilla Linear Regression model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
model1 = LinearRegression()
model1.fit(X_train, y_train)
```

```python
print("Train Performance")
y_pred_train = model1.predict(X_train)
metrics(y_train, y_pred_train)

print("Test Performance")
y_pred_test = model1.predict(X_test)
metrics(y_test, y_pred_test)
```

```python
sns.distplot(y_test - y_pred_test)
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
model2 = Ridge(alpha = 100)
model2.fit(X_train, y_train)

print("Train Performance")
y_pred_train = model2.predict(X_train)
metrics(y_train, y_pred_train)

print("Test Performance")
y_pred_test = model2.predict(X_test)
metrics(y_test, y_pred_test)
```

```python
sns.distplot(y_test - y_pred_test)
```

### Lasso Regression

### Grid Search CV using Ridge

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'alpha':[1,2,3,5,10,15,20,50,100,200]}
grid_model = GridSearchCV(Ridge(), params, cv = 5, scoring = 'r2')

grid_model.fit(X_train, y_train)

print("Train Performance")
y_pred_train = grid_model.predict(X_train)
metrics(y_train, y_pred_train)

print("Test Performance")
y_pred_test = grid_model.predict(X_test)
metrics(y_test, y_pred_test)
```
